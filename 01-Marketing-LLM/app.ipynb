### Understanding chains and Output Parser

```python
from langchain_core.output_parsers import StrOutputParser

prompt = "Hello! who are you?"

template = ChatPromptTemplate.from_messages([
  ('system', 'You are a professional writer'),
  ('human', '{prompt}')
])

chain = template | llm | StrOutputParser()

res = chain.invoke({"prompt":prompt})
res
```

### Improving result exibition

```python
def show_res(res):
  from IPython.display import Markdown
  display(Markdown(res))
  
show_res(res)
```

### Asembling into a function

```python
def llm_generate(llm, prompt):
  template = ChatPromptTemplate.from_messages([
    ('system', 'You are a professional writer'),
    ('human', '{prompt}')
  ])
  
  chain = template | llm | StrOutputParser()
  
  res = chain.invoke({'prompt': prompt})
  show_res(res)
  
```

## 6. Final App

```python
def llm_generate(llm, prompt):
  template = ChatPromptTemplate.from_messages([
      ("system", "You are a digital marketing specialist with focus on SEO and Copywriting."),
      ("human", "{prompt}"),
  ])

  chain = template | llm | StrOutputParser()

  res = chain.invoke({"prompt": prompt})
  return res
```

```python
def generate_result(b):
  with output:
    output.clear_output()
    prompt = f"""
    Write an SEO-optimized text about the topic '{topic.value}'.
    Return only the final text in your response and do not enclose it in quotation marks.
    - Where it will be published: {platform.value}.
    - Tone: {tone.value}.
    - Target audience: {audience.value}.
    - Length: {length.value}.
    - {"Include a clear call to action." if cta.value else "Do not include a call to action."}
    - {"Include relevant hashtags at the end of the text." if hashtags.value else "Do not include hashtags."}
    {"- Keywords that must be included in this text (for SEO): " +  keywords.value if keywords.value else ""}
    """
    try:
      res = llm_generate(llm, prompt)
      show_res(res)
    except Exception as e:
      print(f"Erro: {e}")
```

```python
output = widgets.Output()
generate_button = widgets.Button(description = "Create content")
generate_button.on_click(generate_result)
form = create_form()
```
